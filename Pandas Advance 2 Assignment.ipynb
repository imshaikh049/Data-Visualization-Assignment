```python
Q1. Write a code to print the data present in the second row of the DataFrame, df.


import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Print the second row of the DataFrame
print(df.iloc[1])
```

```python
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
loc: Accesses a group of rows and columns by labels or a boolean array. It is label-based.
Example: df.loc[1] accesses the row with index label 1.
iloc: Accesses a group of rows and columns by integer positions (indices). It is position-based.
Example: df.iloc[1] accesses the second row (index position 1).
```

```python
Q3. Reindex the given DataFrame using a variable, reindex = [3, 0, 1, 2], and store it in the variable, new_df, then find the output for both new_df.loc[2] and new_df.iloc[2].


import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Reindex the DataFrame
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

# Output for new_df.loc[2] and new_df.iloc[2]
print("new_df.loc[2]:")
print(new_df.loc[2])

print("\nnew_df.iloc[2]:")
print(new_df.iloc[2])
Output:

less

new_df.loc[2]:
course_name    Big Data
duration             6
Name: 2, dtype: object

new_df.iloc[2]:
course_name    Data Engineer
duration                 4
Name: 2, dtype: object
```

```python
Q4. Write a code to find the following statistical measurements for the above DataFrame df:
(i) Mean of each column. (ii) Standard deviation of column duration.



import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Mean of each column
mean_values = df.mean(numeric_only=True)
print("Mean of each column:")
print(mean_values)

# Standard deviation of column 'duration'
std_dev_duration = df['duration'].std()
print("\nStandard deviation of 'duration':")
print(std_dev_duration)
```

```python
Q5. Replace the data present in the second row of column, duration, by a string variable then find the mean of column, duration.


import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Replace the data in the second row of 'duration' with a string
df.loc[1, 'duration'] = 'not_a_number'

# Attempt to find the mean of 'duration'
try:
    mean_duration = df['duration'].mean()
except Exception as e:
    mean_duration = str(e)

print("Mean of 'duration':")
print(mean_duration)
Explanation: The mean calculation will fail because the column contains a string, which cannot be converted to a numeric type for the mean calculation.
```

```python
Q6. What do you understand about the windows function in pandas and list the types of window functions?
Windows functions in pandas provide methods to perform operations over a sliding window of data. These functions are used for tasks like rolling calculations, expanding calculations, and exponentially weighted calculations.

Types of window functions:

Rolling: Applies a function to a rolling window of data.
Example: rolling(window=3).mean()
Expanding: Applies a function to an expanding window of data.
Example: expanding().sum()
EWM (Exponentially Weighted Moving): Applies a function with exponentially decreasing weights.
Example: ewm(span=2).mean()
```

```python
Q8. Write a  program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta.


import pandas as pd

# Prompt user for input dates
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Convert input dates to datetime
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# Calculate difference
delta = date2 - date1

# Extract difference in days, hours, and minutes
days = delta.days
hours = delta.components.hours
minutes = delta.components.minutes

print(f"Difference: {days} days, {hours} hours, {minutes} minutes")
```

```python
Q9. Write a  program that reads a CSV file containing categorical data and converts a specified column to a categorical data type.


import pandas as pd

# Prompt user for file path, column name, and category order
file_path = input("Enter the file path: ")
column_name = input("Enter the column name to convert: ")
category_order = input("Enter the category order (comma separated): ").split(',')

# Read CSV file
df = pd.read_csv(file_path)

# Convert specified column to categorical data type
df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

# Display the sorted data
print(df.sort_values(by=column_name))
```

```python
Q10. Write a  program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart.


import pandas as pd
import matplotlib.pyplot as plt

# Prompt user for file path
file_path = input("Enter the file path: ")

# Read CSV file
df = pd.read_csv(file_path)

# Assuming the CSV file has columns 'Date' and 'ProductCategory'
pivot_df = df.pivot_table(index='Date', columns='ProductCategory', values='Sales', aggfunc='sum')

# Plot stacked bar chart
pivot_df.plot(kind='bar', stacked=True)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales by Product Category Over Time')
plt.legend(title='Product Category')
plt.show()
```
